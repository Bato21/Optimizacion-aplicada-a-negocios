### Ejercicio 6

La empresa Chinchulancha, debe decidir cuantos botes de vela debe producir durante los cuatro trimestres del año. La demanda, que debe cumplir en su integridad para mantener su imagen de eficiencia, es durante cada uno de los cuatro trimestres es la que sigue:

| Periodo | Demanda botes de vela |
|:-------:|:---------------------:|
| 1 | 60 |
| 2 | 30 |
| 3 | 50 |
| 4 | 90 |

Al comienzo del primer trimestre, Chinculancha tenía un inventario de 5 botes. La empresa debe decidir cuantos botes producirá durante cada trimestre. En cada trimestre se pueden producir hasta 40 botes con mano de obra regular, con un costo total de \\$400 dólares por bote, pero se puede producir más botes con mano de obra extra, en este caso, el costo del bote es de \\$450 dólares.

Cada bote que se mantenga en el inventario tiene un costo de \\$20 dólares por trimestre. Modele un calendario de producción con el fin de minimizar los costos de producción e inventario durante los cuatro trimestres.

### Solución

- Parámetros:

    - $t$ : [1,2,3,4] 
    - $d_t$ : [60, 30, 50, 90]

- Variables de decisión (todas son variable entera):

    - $r_t$ : Cantidad de botes a producir en el periodo $t$ con mano de obra regular. ($t = 1..4$)
    - $e_t$ : Cantidad de botes a producir en el periodo $t$ con mano de obra extra.
    - $i_t$ : Inventario al final del periodo $t$.

- Restricciones

    - Producción, Demanda, Inventario
 
        $$ i_0 = 5 $$

        $$ i_t = i_{t-1} + r_t + e_t - d_t $$

    - Producción mano de obra regular

        $$ r_t \le 40  $$
    
    - No negatividad

- Función Objetivo

$$ Min \quad Costos = \sum_{t=1}^4 400 r_t + 450 e_t + 20 i_t$$

Nota: Es logistico también.

### Ejercicio 7

Una empresa energética chilena dispone de cuatro plantas de generación para satisfacer la demanda diaria eléctrica en cuatro ciudades, Santiago, Valparaíso, Concepción y Talca. Las plantas 1,2,3 y 4 pueden satisfacer 80, 30, 60 y 45 millones de KW al día respectivamente. Las necesidades de las ciudades de Santiago, Valparaíso, Concepción y Talca son de 70, 40, 70 y 35 millones de Kw al día respectivamente. Los costos asociados al envío de suministro energético por cada millón de KW entre cada planta y cada ciudad son los registrados en la siguiente tabla.


|  | Santiago | Valparaíso | Concepción | Talca |
|:--|:---:|:---:|:---:|:---:|
| Planta 1 | 5 | 2 | 7 | 3 |
| Planta 2 | 3 | 6 | 6 | 1 |
| Planta 3 | 6 | 1 | 2 | 4 |
| Planta 4 | 4 | 3 | 6 | 6 |

Formule un modelo de programación lineal que permita satisfacer las necesidades de todas las ciudades al tiempo que minimice los costos asociados al transporte.
