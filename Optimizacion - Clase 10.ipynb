### Ejercicio 5

Una refinería compra 4 tipos de gasolinas no refinadas con las cuales puede fabricar hasta 3 tipos de combustibles para venta al público.

Los datos de las gasolinas posibles de comprar son las siguientes:

| Tipo | Octanos | Max. Barriles díarios | Precio |
|:----:|:----:|:----:|:----:|
| 1 | 68 | 4000 | 23 |
| 2 | 86 | 5050 | 26 |
| 3 | 91 | 7100 | 28 |
| 4 | 99 | 4300 | 31 |

Se puede producir gasolina refinada mezclando los productos comprados de acuerdo a lo siguiente información:

| Tipo | Min Octanos | Barriles díarios | Precio |
|:----:|:----:|:----:|:----:|
| 1 | 95 | Hasta 10000 | 45 |
| 2 | 90 | Cualquier cantidad | 43 |
| 3 | 85 | Minimo 15000 | 40 |

Adicionalmente, la refinería puede vender gasolina no refinada a \\$39 si tiene un octanaje mayor a 90 octanos y \\$37 si el octanaje es menor a 90 octanos.

**Modele el problema de programación lineal para maximizar las ganancias de diarias de la refinería**.

#### Solución

a. Variables de Decisión:

- $x_i$: Barriles de Gasolina tipo i vendidas sin refinar ($i=1..4$)

- $x_{ij}$: Barriles de Gasolina tipo i usados para fabricar tipo j ($i=1..4$, $j=1..3$)

b. Restricciones:

- Disponibilidad barriles diarios:

$$ x_1 + x_{11} + x_{12} + x_{13} \le 4000 $$

$$ x_2 + x_{21} + x_{22} + x_{23} \le 5050 $$

$$ x_3 + x_{31} + x_{32} + x_{33} \le 7100 $$

$$ x_4 + x_{41} + x_{42} + x_{43} \le 4300 $$

- Octanaje mínimo (promedio ponderado):

$$ \frac{68 x_{11} + 86 x_{21} + 91 x_{31} + 99 x_{41}}{x_{11} + x_{21} + x_{31} + x_{41}} \ge 95 $$

$$ \frac{68 x_{12} + 86 x_{22} + 91 x_{32} + 99 x_{42}}{x_{12} + x_{22} + x_{32} + x_{42}} \ge 90 $$

$$ \frac{68 x_{13} + 86 x_{23} + 91 x_{33} + 99 x_{43}}{x_{13} + x_{23} + x_{33} + x_{43}} \ge 85 $$

- Límites de venta:

$$ x_{11} + x_{21} + x_{31} + x_{41} \le 10000 $$

$$ x_{13} + x_{23} + x_{33} + x_{43} \ge 15000 $$

- No negatividad, las de siempre.

c. Función objetivo

$$ Máx (Ingresos - Costos)  $$

- $Ingresos$ (Producción):

$$ 45 (x_{11}+x_{21}+x_{31}+x_{41}) + 43 (x_{12}+x_{22}+x_{32}+x_{42}) + 40 (x_{13}+x_{23}+x_{33}+x_{43}) $$

- $Ingresos$ (Intermediación):

$$ 39 (x_3+x_4) + 37 (x_1+x_2) $$

- $Costos$:

$$ 23 (x_1+x_{11}+x_{12}+x_{13}) + 26 (x_2+x_{21}+x_{22}+x_{23}) + 28 (x_3+x_{31}+x_{32}+x_{33}) + 31 (x_4+x_{41}+x_{42}+x_{43})$$

```python
from scipy.optimize import minimize
import numpy as np

def fo(x):
    pcompra = np.array([23,23,23,23,26,26,26,26,28,28,28,28,31,31,31,31])
    pventa = np.array([37,45,43,40,37,45,43,40,39,45,43,40,39,45,43,40])
    return pcompra@x - pventa@x

r1 = lambda x: -1*(sum(x[:4]) - 4000)
r2 = lambda x: -1*(sum(x[4:8]) - 5050)
r3 = lambda x: -1*(sum(x[8:12]) - 7100)
r4 = lambda x: -1*(sum(x[12:]) - 4300)

def r5(x):
    oct = np.array([68,86,91,99])
    return (95 - (oct@x[1::4])/sum(x[1::4]))
def r6(x):
    oct = np.array([68,86,91,99])
    return (90 - (oct@x[2::4])/sum(x[2::4]))
def r7(x):
    oct = np.array([68,86,91,99])
    return (85 - (oct@x[3::4])/sum(x[3::4]))

r8 = lambda x: -1*(sum(x[1::4]) - 10000)
r9 = lambda x: -1*(15000 - sum(x[3::4]))

# si hay mas restricciones se agregan a la lista
rest_list = [{'type': 'ineq', 'fun': r1}, {'type': 'ineq', 'fun': r2}, {'type': 'ineq', 'fun': r3}, \
             {'type': 'ineq', 'fun': r4}, {'type': 'ineq', 'fun': r5}, {'type': 'ineq', 'fun': r6}, \
             {'type': 'ineq', 'fun': r7}, {'type': 'ineq', 'fun': r8}, {'type': 'ineq', 'fun': r9}]

# las nunca bien ponderadas restricciones de no negatividad
noneg = [(0, 500000)]*16

solucion_inicial = np.ones(16)

# Minimize the function using the Nelder-Mead method
result = minimize(fo, solucion_inicial, method='SLSQP', bounds=noneg, constraints=rest_list)

# Print the result
print(result)
```
