### Optimización con restricciones

Ahora vamos a agregarle a nuestro problema de optimización restricciones de desigualdad y resolverlo.

#### Ejercicio 1 (mismo clase anterior)

$$ min \left( -x - 2 y \right) $$

Con $x \geq 0$ e $y \geq 0$, sujeto a:

$$ R_1: 2 x + 6 y \leq 36 $$ 
$$ R_2: 5 x + 3 y \leq 30$$
$$ R_3: 8 x + 2 y \leq 40$$

Vamos paso a paso

$$ L(x,y,\lambda_1,\lambda_2,\lambda_3) = -x -2y + \lambda_1 (2x+6y-36) + \lambda_2 (5x+3y-30) + \lambda_3 (8x+2y-40)$$

$$ \frac{\partial L}{\partial x} : -1 + 2\lambda_1 + 5\lambda_2 + 8\lambda_3 = 0 $$

$$ \frac{\partial L}{\partial y} : -2 + 6\lambda_1 + 3\lambda_2 + 2\lambda_3 = 0 $$

$$ \frac{\partial L}{\partial \lambda_1} : 2x + 6y -36 = 0 $$

$$ \frac{\partial L}{\partial \lambda_1} : 5x + 3y -30 = 0 $$

$$ \frac{\partial L}{\partial \lambda_1} : 8x + 2y -40 = 0 $$

Se puede expresar en el siguiente sistema de ecuaciones lineal.

$$ \left[
  \begin{array}{ccccc}
    0 & 0 & 2 & 5 & 8 \\
    0 & 0 & 6 & 3 & 2 \\
    2 & 6 & 0 & 0 & 0 \\
    5 & 3 & 0 & 0 & 0 \\
    8 & 2 & 0 & 0 & 0 \\
  \end{array}
\right]
\begin{bmatrix}
           x \\
           y \\
           \lambda_1 \\
           \lambda_2 \\
           \lambda_3 \\
         \end{bmatrix}
= \begin{bmatrix}
           1 \\
           2 \\
           36 \\
           30 \\
           40 \\
         \end{bmatrix}
$$

Pero, las restricciones me entregan 3 ecuaciones para 2 variables, por lo tanto no son linealmente independientes, entonces para resolver esto debe eliminar la restriccion que no es activa. Eliminando las tercera restricción:

$$ \left[
  \begin{array}{ccccc}
    0 & 0 & 2 & 5 \\
    0 & 0 & 6 & 3 \\
    2 & 6 & 0 & 0 \\
    5 & 3 & 0 & 0 \\
  \end{array}
\right]
\begin{bmatrix}
           x \\
           y \\
           \lambda_1 \\
           \lambda_2 \\
         \end{bmatrix}
= \begin{bmatrix}
           1 \\
           2 \\
           36 \\
           30 \\
         \end{bmatrix}
$$


```python
import numpy as np

A = np.array([[0, 0, 2, 5],
              [0, 0, 6, 3],
              [2, 6, 0, 0],
              [5, 3, 0, 0]])

b = np.array([1, 2, 36, 30])

# Calculating the inverse of the matrix
Ainv = np.linalg.inv(A)

sol = Ainv @ b
sol
```

La coordenada (3,5) es la que nos mostro la solución gráfica. Otra forma:

```python
from scipy.optimize import minimize
import numpy as np

fo = lambda x: -x[0]-2*x[1]
r1 = lambda x: -2*x[0] - 6*x[1] + 36
r2 = lambda x: -5*x[0] - 3*x[1] + 30
r3 = lambda x: -8*x[0] - 2*x[1] + 40

# si hay mas restricciones se agregan a la lista
rest_list = [{'type': 'ineq', 'fun': r1}, {'type': 'ineq', 'fun': r2}, {'type': 'ineq', 'fun': r3}]

# las nunca bien ponderadas restricciones de no negatividad
noneg = [(0, np.inf), (0, np.inf)]

solucion_inicial = np.array([0, 0])

# Minimize the function using the Nelder-Mead method
result = minimize(fo, solucion_inicial, method='SLSQP', bounds=noneg, constraints=rest_list)

# Print the result
print(result)
```
