```python
import gurobipy as gp
from gurobipy import GRB

try:
    # Esto inicializa el entorno y verifica tu licencia automáticamente
    modelo = gp.Model("Prueba_WLS")

    # Crear variables simples
    x = modelo.addVar(vtype=GRB.CONTINUOUS, name="x")
    y = modelo.addVar(vtype=GRB.CONTINUOUS, name="y")

    # Función objetivo: Maximizar x + y
    modelo.setObjective(x + y, GRB.MAXIMIZE)

    # Añadir una restricción: x + 2y <= 10
    modelo.addConstr(x + 2 * y <= 10, "restriccion_1")
    
    # Resolver el modelo
    modelo.optimize()

    print("\n--- ¡Licencia funcionando correctamente! ---")
    for v in modelo.getVars():
        print(f"{v.varName}: {v.X}")
    print(f"Valor objetivo: {modelo.objVal}")

except gp.GurobiError as e:
    print(f"Error de Gurobi. Código: {e.errno}. Descripción: {e}")
```
