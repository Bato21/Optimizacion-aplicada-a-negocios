### Condiciones KKT

Se aplica en problemas del tipo:

$$ mín f(x) $$

sujeto a,

$$ g_i(x) \le 0 , \quad i = 1, ..., n$$
$$ h_j(x) = 0 , \quad j = 1, ..., m$$


Si, $f, \quad g_i \quad y \quad h_i$ son continuas y diferenciables, si existe un minimo local $x^*$, entonces existen los multiplicadores KKT donde  $\mu_i \ge 0, i=1,...,n$ y existen reales $\lambda_j, j=1,...,m$. Si esto se cumple, entonces:

- Condiciones de estacionariedad

$$ \nabla f(x^*) + \sum_{i=1}^n \mu_i \nabla g_i(x^*) + \sum_{j=1}^m \lambda_j h_j(x^*) = 0$$

- Condiciones de holgura complementaria

$$ \mu_i g_i(x^*) = 0 , \quad \quad \forall i = 1,...,n $$

- Condiciones de factibilidad primal

$$ g_i(x^∗) \le 0 , \quad \quad \forall i = 1,...,n $$

$$h_j(x^∗) = 0 , \quad \quad \forall j = 1,...,m$$

- Condiciones regularidad:

  - Si $g_i$ y $h_i$ son funciones afines (lineales) existe $x^*$
 
  - Los gradientes de las restricciones de desigualdad activas más los gradientes de las restricciones de igualdad, son linealmente independientes.
 
  - **Condición de slater**: Para un problema convexo, es decir $f$ y $g_i$ convexa y $h_i$ afin, entonces existe $x^*$ tal que $h_i(x^*) = 0$ y $g_i(x^*) \lt 0$.
 
En resumen, si la función objetivo $f$ es una función convexa, las restricciónes de desigualdad son funciones diferenciables
y convexas y las restricciones de igualdad son funciones afines. Entonces, las condiciones de KKT son suficientes para garantizar la optimalidad de un punto.

#### Ejercicio 1.

Considere el siguiente problema de optimización no lineal P.

$$ máx f(x,y) = 2x-y $$

$$ 3x+2y \le 20 $$

$$ x+y \ge 4$$

$$ x^2 -4x - 2y \ge -8 $$ 

- Encuentre todos los minimos locales de P y verifique que cumplen con las condiciones de KKT.
- Para cada punto encontrado, muestre gráficamente la condición.

#### Solución 1.

Expresar P en forma estándar.

$$ mín f(x,y) = -2x + y $$

$$ g_0: 3x + 2y -20 \le 0$$

$$ g_1: -x - y + 4 \le 0$$

$$ g_2: -x^2 + 4x + 2y -8 \le 0$$

Calculando el gradiente de cada expresión:

$$ \nabla f = \left( \begin{array}{c} -2 \\ 1 \end{array} \right)$$

$$ \nabla g_0 = \left( \begin{array}{c} 3 \\ 2 \end{array} \right)$$

$$ \nabla g_1 = \left( \begin{array}{c} -1 \\ -1 \end{array} \right)$$

$$ \nabla g_2 = \left( \begin{array}{c} -2x+4 \\ 2 \end{array} \right)$$

Se puede probando, 0, 1, 2 o las 3 restricciones activas, 8 casos. Pero veamos gráficamente.

```python
import matplotlib.pyplot as plt
import numpy as np

g0 = lambda x: (-3*x+20)/2
g1 = lambda x: -x+4
g2 = lambda x: (x**2-4*x+8)/2

obj = lambda x,c: (2*x+c)

x = np.linspace(-10, 20)
plt.ylim(-10,20)
plt.plot(x, g0(x))
plt.plot(x, g1(x))
plt.plot(x, g2(x))
plt.fill_between(x,-10,g0(x), color="red", alpha=0.25)
plt.fill_between(x,20,g1(x), color="red", alpha=0.25)
plt.fill_between(x,-10,g2(x), color="red", alpha=0.25)

xpoints = [12,0,2,4,-3]
ypoints = [-8,4,2,4,14.5]
plt.scatter(xpoints, ypoints, color="black")

for i in range(10):
    plt.plot(x, obj(x,i), color="brown", alpha=(1-(i/12)))
plt.show()
```

Graficamente, las restricciones activas son $g_0$ y $g_1$ en el punto $(12,-8)$, luego solo hay que verificar las condiciones de KKT en este punto, es decir encontrar $\mu_0, \mu_1, \mu_2 \ge 0$, reemplazando se obtiene que:

$$ \mu_0 = 3, \quad \mu_1 = 7, \quad \mu_2 = 0 $$

Encontramos el minimo global, ¿existiran minimos locales?, veamos otros casos.

Caso $g_1$ y $g_2$ activas y $g_0$ inactiva, si resolvemos el sistema de ambas, se encuentra los puntos $(0,4)$ y $(2,2)$.

La otra variante entrega los puntos $(4,4)$ y $(-3, 14.5)$, en todos estos casos ver si existe $\mu_i \gt 0$, para determinar si es minimo local.
