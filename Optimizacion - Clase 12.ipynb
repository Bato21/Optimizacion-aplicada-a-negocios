Ejercicios modelamiento lineal, otros tipos de problemas.

Problemas finanzas.

### Ejercicio 9

Un inversionista dispone de un capital limitado para invertir en $I$ posibles activos de renta variable (acciones y bonos). Se estima que el el retorno esperado del activo $i$ es $r_i$. Por otro lado el riesgo de cada activo $i$ se ha estimado en $\theta_i$. Considere que el retorno y el riesgo tienen un comportamiento lineal con la inversión. Construya un modelo de programación lineal que permita determinar como repartir la inversión de manera de tener el mínimo riesgo posible, asegurando un retorno mínimo R.

Problemas de finanzas multiperiodo.

### Ejercicio 10

Una compañía de inversiones dispone de un capital inicial igual a $K$ millones para invertir a lo largo de $T$ meses de un año. Considere que existe un total de $M$ instrumentos de inversión, los que al final del mes, entregan una cierta rentabilidad estrictamente positiva por cada peso invertido al comienzo de dicho mes.

Sea $\alpha_{it}$ la rentabilidad que se obtiene al final del mes $t$ por invertir un peso en el instrumento $i$ al comienzo de dicho mes (de esta forma, si $\alpha_{it} = 3% $ y usted invierte \\$100 en el instrumento $i$, al final de dicho mes tendría \\$103, los que podría reinvertir en alguno de los $M$ instrumentos durante el mes $t + 1$).

En cada mes, debe decidirse en cuáles instrumentos invertir el dinero disponible al comienzo de dicho mes (es decir, las ganancias obtenidas por sus decisiones de inversión realizadas en el mes anterior) y en qué cantidad. Por último, considere que una estrategia racional la constituye invertir al comienzo del primer mes todo su capital inicial, por cuanto $\alpha_{it} > 0$, $\forall i, t$.

a. Escriba un modelo de optimización que permita determinar el plan óptimo de inversión, para el caso donde T = 2. ¿Qué función objetivo le permitiría maximizar su ganancia final?

b. Escriba un modelo de optimización que permita determinar el plan óptimo de inversión, es decir, aquel que maximiza los retornos, para el caso donde $T = 12$. ¿Cómo cree usted que es la solución óptima de este problema?

Problemas optimización lineal mixto (variables continua y binaria)

### Ejercicio 11

Una empresa europea piensa instalar plantas de producción en Chile para lanzar sus productos al mercado chileno por lo que necesita decidir su plan de producción para el próximo ano. La empresa puede fabricar $N$ productos distintos y la elaboración de cada uno de ellos implica la compra de una máquina especializada para su elaboración a un costo de $f_n$.

Además, el costo variable de producir una unidad del producto $n$ es de $c_n$. Así, si se decide elaborar el producto $n$ se deberá necesariamente incurrir en un costo de $f_n$ más los costos variables por elaboración del producto y si se decide no fabricarlo no se incurrirá en ningún tipo de gasto.

Si la demanda pronosticada para el producto $n$ es de $D_n$ unidades (n=1...N) pudiendo venderse dicho producto a un precio de $p_n$, formule un PPL mixto que resuelva el problema de encontrar el conjunto de productos que la empresa debe fabricar.
